```python
import os, subprocess
files = []
for filename in os.listdir('/home/Aghori/Downloads/Resume_Data'):
    '''The Resumes are stored in the above location'''
    if filename.endswith('.doc'):
        print(filename)
        subprocess.call(['soffice', '--headless', '--convert-to', 'docx', filename])
        os.remove(filename)
        '''As th format of resumes are in .doc & .docx format is necessary to convert them into the same type.
        It can also be done by converting into text file directly.
        With os.remove after converting the .doc files into .docx have deleted the .doc originals
        Below are the list of .doc files '''
    files.append(filename)
```

```python
import docx
import re
Nam = []
'''Now going through each files extracting the first 4 lines as there is a more chance of getting Name from it'''
for f in files:
    if f.endswith('.docx'):
        doc = docx.Document(f)
        fulltext = []
        for para in doc.paragraphs[0:4]:
            # Each line is a pragraph in the understanding of the docx library
            fulltext.append(para.text)
            fulltext = '\n'.join(fulltext)
            fulltext = re.findall('([A-Z][a-z]+)', fulltext)
            '''The Regex is used in order to filter the data, i.e., extracting only words not numbers etc '''
        Nam.append(fulltext)
```
